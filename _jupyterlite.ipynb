2. **Задача 1: Кластеризация методом K-средних**

Описание: Используйте алгоритм K-средних для кластеризации набора данных Iris

```python
# Импорт библиотек

import numpy as np

import pandas as pd

from sklearn.datasets import load_iris

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt



# Загрузка данных

iris = load_iris()

X = iris.data



# Кластеризация

kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(X)



# Визуализация

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')

plt.title('Кластеризация Iris методом K-средних')

plt.xlabel('Длина чашелистика')

plt.ylabel('Ширина чашелистика')

plt.show()
```

**Вывод**: Алгоритм K-средних успешно разделил данные на три кластера, соответствующих различным видам ирисов.

3. **Задача 2: Снижение размерности методом PCA**

Описание: Примените метод главных компонент (PCA) для снижения размерности данных Iris.

python

from sklearn.decomposition import PCA

```python
from sklearn.decomposition import PCA



# Снижение размерности

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)



# Визуализация

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')

plt.title('Снижение размерности Iris методом PCA')

plt.xlabel('Первая главная компонента')

plt.ylabel('Вторая главная компонента')

plt.show()
```

**Вывод**: PCA позволил уменьшить размерность данных с сохранением основных закономерностей.

3. **Поиск ассоциативных правил**

Задача: Поиск взаимосвязей между переменными.
Пример: Используйте алгоритм Apriori для анализа покупок в интернет-магазине.

```python
from mlxtend.frequent_patterns import apriori

from mlxtend.frequent_patterns import association_rules

import pandas as pd



# Пример данных

data = {'Товар 1': [1, 0, 1, 1], 'Товар 2': [1, 1, 0, 1], 'Товар 3': [0, 1, 1, 0]}

df = pd.DataFrame(data)



# Поиск частых наборов

frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

print(rules)
```

**Вывод**: Алгоритм выявил ассоциативные правила, полезные для маркетинга35.

4. **Обнаружение аномалий**

Задача: Определение необычных или аномальных паттернов в данных.
Пример: Используйте метод изолированных лесов (Isolation Forest) для обнаружения мошеннических транзакций.

```python
from sklearn.ensemble import IsolationForest

import numpy as np



# Пример данных

X = [[-1.1], [0.3], [0.5], [100]]

clf = IsolationForest(random_state=42)

clf.fit(X)



# Обнаружение аномалий

print(clf.predict(X))
```

**Вывод**: Алгоритм успешно идентифицировал аномалии25.

5. **Генеративные модели**

Задача: Создание новых данных на основе существующих.
Пример: Используйте автоэнкодеры для генерации изображений рукописных цифр.

```python
from keras.layers import Input, Dense

from keras.models import Model



# Пример автоэнкодера

input_img = Input(shape=(784,))

encoded = Dense(32, activation='relu')(input_img)

decoded = Dense(784, activation='sigmoid')(encoded)

autoencoder = Model(input_img, decoded)

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

**Вывод:** Автоэнкодер успешно воспроизводит данные48.

1. **Обнаружение мошеннических транзакций**

Описание: Используйте данные о банковских транзакциях для выявления подозрительных операций.
Методы: One-Class SVM, Isolation Forest.

```python
from sklearn.ensemble import IsolationForest

import pandas as pd



# Пример данных

data = {'Сумма': [100, 200, 300, 10000], 'Время': [10, 20, 30, 5]}

df = pd.DataFrame(data)



# Обнаружение аномалий

clf = IsolationForest(random_state=42)

df['Аномалия'] = clf.fit_predict(df)

print(df)
```

2. **Обнаружение сбоев в работе механизмов**

3. **Обнаружение аномалий во временных рядах**

Описание: Используйте данные временных рядов для выявления необычных паттернов.
Методы: Прогнозирование временных рядов, PCA.
Код:

```python
from sklearn.decomposition import PCA

import numpy as np



# Пример данных

X = np.random.rand(100, 5)

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)



# Визуализация

import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0], X_pca[:, 1])

plt.show()
```

**Вывод**: PCA помог выявить аномалии в данных35.

4. **Обнаружение аномалий в графах**

Описание: Анализируйте графы для выявления необычных узлов или связей.
Методы: Метрические методы, кластеризация

```python
from sklearn.neighbors import LocalOutlierFactor



# Пример данных

X = [[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]]

clf = LocalOutlierFactor(n_neighbors=2)

clf.fit_predict(X)
```

**Вывод**: Метод успешно определил аномальные узлы в графе13.
